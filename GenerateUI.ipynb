```python
from pathlib import Path
from datetime import datetime
import base64
import mimetypes
import re, json
import pandas as pd

ruta_index = Path(r"C:\Users\Jorge Vasquez\DocoOP\index.html")
ruta_base = ruta_index.parent
ruta_passwords = ruta_base / "SSWO.xlsx"
ruta_ranking = Path(r"C:\Users\Jorge Vasquez\Ranking\CS_AVANCE DE ASESORES.xlsx")

MESES_HOJAS = [
    "ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
    "JULIO", "AGOSTO", "SETIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"
]

def norm_text(v):
    return "" if pd.isna(v) else str(v).strip()

def norm_key(v):
    return re.sub(r"\s+", " ", norm_text(v).upper())

def hoja_mes_actual(hoy=None):
    hoy = hoy or datetime.now()
    return f"{MESES_HOJAS[hoy.month - 1]}_{hoy.year}"

def clave_dim_mes(hoy=None):
    hoy = hoy or datetime.now()
    return f"{hoy.year}_{hoy.month}"

def cargar_passwords(ruta_excel: Path):
    if not ruta_excel.exists():
        raise FileNotFoundError(f"No se encontró el Excel de passwords: {ruta_excel}")

    df = pd.read_excel(ruta_excel, sheet_name="PASSWORD", dtype=str)
    df.columns = [str(c).strip().upper() for c in df.columns]

    columnas_requeridas = ["USUARIO", "CLAVE"]
    faltantes = [c for c in columnas_requeridas if c not in df.columns]
    if faltantes:
        raise ValueError(f"Faltan columnas en hoja PASSWORD: {faltantes}")

    for col in df.columns:
        df[col] = df[col].map(norm_text)

    passwords = {}
    for _, row in df.iterrows():
        usuario = row["USUARIO"].strip()
        clave = row["CLAVE"].strip()
        if usuario:
            passwords[usuario] = clave

    return passwords

PASSWORDS = cargar_passwords(ruta_passwords)
PASSWORDS_JSON = json.dumps(PASSWORDS, ensure_ascii=False)

ENTIDAD_OPTIONS = [
    "Caja Municipal de Ahorro y Crédito de Arequipa S.A.",
    "Caja Municipal de Ahorro y Crédito de Trujillo S.A.",
    "Banco de la Nación",
    "Banco BBVA Perú",
    "Financiera Credinka S.A.",
    "Caja Rural de Ahorro y Crédito del Centro S.A.",
    "Banco Internacional del Perú S.A.A. – Interbank",
    "Caja Rural de Ahorro y Crédito Los Andes S.A.",
    "Mibanco – Banco de la Microempresa S.A.",
    "Financiera ProEmpresa S.A.",
    "Financiera Qaqap S.A.C.",
    "Caja Municipal de Ahorro y Crédito de Sullana S.A.",
    "Banco Scotiabank Perú S.A.A.",
]

def imagen_a_data_uri(ruta_img: Path) -> str:
    if not ruta_img.exists():
        raise FileNotFoundError(f"No se encontró la imagen: {ruta_img}")
    mime, _ = mimetypes.guess_type(str(ruta_img))
    if not mime:
        mime = "application/octet-stream"
    data = base64.b64encode(ruta_img.read_bytes()).decode("utf-8")
    return f"data:{mime};base64,{data}"

LOGO_DATA_URI = imagen_a_data_uri(ruta_base / "images" / "LOGO_home.jpg")
FIRMA_NATHALY_DATA_URI = imagen_a_data_uri(ruta_base / "images" / "FIRMA_nathaly.jpg")

CARTA_CDNA = CARTA_CDNA.replace('src="images/LOGO_home.jpg"', f'src="{LOGO_DATA_URI}"')
CARTA_CDNA = CARTA_CDNA.replace('src="images/FIRMA_nathaly.jpg"', f'src="{FIRMA_NATHALY_DATA_URI}"')
CARTA_AMONESTACION = CARTA_AMONESTACION.replace(
    'src="images/LOGO_home.jpg"',
    f'src="{LOGO_DATA_URI}"',
)

def extraer_dia_fecha(v):
    fecha = parse_ind_fecha(v)
    if pd.isna(fecha):
        return ""
    return f"{int(fecha.day):02d}"

def es_si(v):
    return norm_key(v) == "SI"

def parse_ind_fecha(v):
    raw = norm_text(v)
    if not raw:
        return pd.NaT

    iso = re.match(r"^(\d{4})[-/](\d{1,2})[-/](\d{1,2})", raw)
    if iso:
        return pd.Timestamp(year=int(iso.group(1)), month=int(iso.group(2)), day=int(iso.group(3)))

    local = re.match(r"^(\d{1,2})[-/](\d{1,2})[-/](\d{2,4})", raw)
    if local:
        year = int(local.group(3))
        if year < 100:
            year += 2000
        return pd.Timestamp(year=year, month=int(local.group(2)), day=int(local.group(1)))

    return pd.to_datetime(raw, errors="coerce", dayfirst=True)

def fecha_iso_from_ind(v):
    fecha = parse_ind_fecha(v)
    if pd.isna(fecha):
        return ""
    return fecha.strftime("%Y-%m-%d")

def fecha_larga_es_from_ind(v):
    fecha = parse_ind_fecha(v)
    if pd.isna(fecha):
        return ""
    return f"{int(fecha.day)} de {MESES_HOJAS[int(fecha.month) - 1].capitalize()} de {int(fecha.year)}"

def cargar_indicadores_mes(ruta_excel: Path, hoja: str):
    raw = pd.read_excel(ruta_excel, sheet_name=hoja, header=None, dtype=str)
    posiciones = [
        (fila, col)
        for fila in range(raw.shape[0])
        for col in range(raw.shape[1])
        if norm_text(raw.iat[fila, col]) == "IND_Fecha"
    ]
    if not posiciones:
        raise ValueError(f"No se encontró la columna IND_Fecha en la hoja {hoja}")

    fila_header, _ = posiciones[0]
    df = pd.read_excel(ruta_excel, sheet_name=hoja, header=fila_header, dtype=str)
    columnas_ind = [c for c in df.columns if str(c).strip().startswith("IND_")]
    df = df[columnas_ind].copy()

    requeridas = ["IND_Fecha", "IND_Alias", "IND_FALTA", "IND_TardanzaTiempo", "IND_TARDE"]
    faltantes = [c for c in requeridas if c not in df.columns]
    if faltantes:
        raise ValueError(f"Faltan columnas en indicadores ({hoja}): {faltantes}")

    for col in df.columns:
        df[col] = df[col].map(norm_text)
    df = df[df["IND_Alias"] != ""]
    return df

def cargar_consolidado_mes(ruta_excel: Path, hoja: str):
    df = pd.read_excel(ruta_excel, sheet_name=hoja, header=1, dtype=str)
    df.columns = [str(c).strip().upper() for c in df.columns]
    for col in df.columns:
        df[col] = df[col].map(norm_text)
    if "ASESOR" not in df.columns:
        return {}
    return {
        norm_key(row["ASESOR"]): row
        for _, row in df.iterrows()
        if norm_text(row.get("ASESOR", ""))
    }

def cargar_dim_mes(ruta_excel: Path, clave_mes: str):
    df = pd.read_excel(ruta_excel, sheet_name="DIM", dtype=str)
    df.columns = [str(c).strip().upper() for c in df.columns]
    requeridas = ["FECHA", "ASESOR", "DNI", "EQUIPO", "CONDICION"]
    faltantes = [c for c in requeridas if c not in df.columns]
    if faltantes:
        raise ValueError(f"Faltan columnas en hoja DIM: {faltantes}")

    for col in df.columns:
        df[col] = df[col].map(norm_text)
    df = df[df["FECHA"].map(norm_key) == norm_key(clave_mes)]
    df = df[df["ASESOR"] != ""].copy()
    df["CONDICION"] = df["CONDICION"].map(norm_key)
    return df

def resumen_indicadores_por_asesor(df_indicadores):
    resumen = {}
    for _, row in df_indicadores.iterrows():
        asesor = norm_key(row["IND_Alias"])
        if not asesor:
            continue
        item = resumen.setdefault(asesor, {"faltas": set(), "tardanzas": set(), "tiempos": {}, "registros": []})
        dia = extraer_dia_fecha(row["IND_Fecha"])
        fecha_iso = fecha_iso_from_ind(row["IND_Fecha"])
        if not dia or not fecha_iso:
            continue

        falta = es_si(row["IND_FALTA"])
        tarde = es_si(row["IND_TARDE"])
        tiempo = row.get("IND_TardanzaTiempo", "")

        item["registros"].append({
            "fecha": fecha_iso,
            "dia": dia,
            "falta": "SI" if falta else "NO",
            "tardanza": "SI" if tarde else "NO",
            "tiempo": tiempo,
        })

        if falta:
            item["faltas"].add(dia)
        if tarde:
            item["tardanzas"].add(dia)
            item["tiempos"][dia] = tiempo

    return {
        asesor: {
            "faltas_dias": sorted(datos["faltas"], key=int),
            "tardanzas_dias": sorted(datos["tardanzas"], key=int),
            "tardanza_tiempos": datos["tiempos"],
            "registros_base": sorted(datos["registros"], key=lambda x: x["fecha"]),
        }
        for asesor, datos in resumen.items()
    }

def cargar_catalogo_amonestacion(ruta_excel: Path):
    if not ruta_excel.exists():
        raise FileNotFoundError(f"No se encontró el Excel de ranking: {ruta_excel}")

    hoja = hoja_mes_actual()
    clave_mes = clave_dim_mes()
    xl = pd.ExcelFile(ruta_excel)
    if hoja not in xl.sheet_names:
        raise ValueError(f"No existe la hoja del mes actual: {hoja}")

    df_indicadores = cargar_indicadores_mes(ruta_excel, hoja)
    fechas_ind = df_indicadores["IND_Fecha"].map(parse_ind_fecha)
    fecha_max = fechas_ind.max()
    fecha_max_iso = "" if pd.isna(fecha_max) else fecha_max.strftime("%Y-%m-%d")
    fecha_max_larga = "" if pd.isna(fecha_max) else fecha_larga_es_from_ind(fecha_max_iso)

    indicadores = resumen_indicadores_por_asesor(df_indicadores)
    consolidado = cargar_consolidado_mes(ruta_excel, hoja)
    dim = cargar_dim_mes(ruta_excel, clave_mes)

    supervisores_dim = dim[dim["CONDICION"] == "SUPERVISOR"].copy()
    supervisor_por_equipo = {}
    for _, row in supervisores_dim.iterrows():
        equipo = norm_key(row["EQUIPO"])
        if equipo and equipo not in supervisor_por_equipo:
            supervisor_por_equipo[equipo] = row["ASESOR"]

    def persona_desde_row(row):
        nombre = row["ASESOR"]
        nombre_key = norm_key(nombre)
        equipo = row["EQUIPO"]
        supervisor = supervisor_por_equipo.get(norm_key(equipo), equipo)
        datos_consolidados = consolidado.get(nombre_key, {})
        dni = row["DNI"] or norm_text(datos_consolidados.get("DNI", ""))
        datos_indicador = indicadores.get(nombre_key, {})
        return {
            "nombre": nombre,
            "dni": dni,
            "supervisor": supervisor,
            "faltas_dias": datos_indicador.get("faltas_dias", []),
            "tardanzas_dias": datos_indicador.get("tardanzas_dias", []),
            "tardanza_tiempos": datos_indicador.get("tardanza_tiempos", {}),
            "registros_base": datos_indicador.get("registros_base", []),
        }

    asesores_call = (
        dim[dim["CONDICION"] == "CALL"]
        .drop_duplicates(subset=["ASESOR"], keep="first")
        .sort_values("ASESOR")
    )
    supervisores = (
        supervisores_dim
        .drop_duplicates(subset=["ASESOR"], keep="first")
        .sort_values("ASESOR")
    )

    return {
        "asesor": [persona_desde_row(row) for _, row in asesores_call.iterrows()],
        "supervisor": [persona_desde_row(row) for _, row in supervisores.iterrows()],
        "excel_usado": ruta_excel.name,
        "hoja_usada": hoja,
        "dim_mes": clave_mes,
        "fecha_max_indicadores": fecha_max_iso,
        "fecha_max_indicadores_larga": fecha_max_larga,
    }

CATALOGO_AMONESTACION = cargar_catalogo_amonestacion(ruta_ranking)
CATALOGO_AMONESTACION_JSON = json.dumps(CATALOGO_AMONESTACION, ensure_ascii=False)

# =========================
# CONFIGURACIÓN DE CARTAS
# =========================
MODELOS = {
"cdna": {
    "nombre": "CARTA DE NO ADEUDO",
    "template": CARTA_CDNA,
    "order": [
        "DOC",
        "NOMBRE",
        "FECHA_PAGO",
        "ENTIDAD",
        "FECHA_CASTIGO",
        "CREDITOS",
    ],
    "labels": {
        "FECHA_HOY": "Fecha de emisión",
        "TIPO_DOC": "Tipo de documento",
        "DOC": "N° de documento",
        "NOMBRE": "Nombre completo",
        "ENTIDAD": "Entidad (banco / empresa)",
        "FECHA_CASTIGO": "Fecha de cesión / castigo",
        "CREDITOS": "Créditos",
        "FECHA_PAGO": "Fecha de pago",
    },
    "date_keys": {"FECHA_HOY", "FECHA_CASTIGO", "FECHA_PAGO"},
    "select_keys": {"ENTIDAD"},
    "textarea_keys": set(),
    "defaults": {},
    "extra_fields_top": """
    <div class="field-wrap" data-scope="all">
      <div class="field">
        <label class="mini-label">Tipo de documento</label>
        <div class="switch-rol" id="cdna_switch_doc">
          <button type="button" class="switch-rol-btn active" data-value="DNI" id="btnDocDni">
            DNI
          </button>
          <button type="button" class="switch-rol-btn" data-value="RUC" id="btnDocRuc">
            RUC
          </button>
        </div>
        <input id="cdna_TIPO_DOC" type="hidden" value="DNI">
      </div>
    </div>
    """.strip(),
    
    "extra_fields_bottom": """
    <div class="field-wrap hidden" data-scope="all" id="cdna_EXPEDIENTES_WRAP">
      <div class="field field-creditos">
        <div class="creditos-header">
          <label>N° de expediente</label>
        </div>
        <div class="creditos-grid" id="cdna_EXPEDIENTES_CONTAINER">
          <input id="cdna_EXPEDIENTE_1" type="text" placeholder="Expediente 1">
          <input id="cdna_EXPEDIENTE_2" type="text" placeholder="Expediente 2">
          <input id="cdna_EXPEDIENTE_3" type="text" placeholder="Expediente 3">
          <input id="cdna_EXPEDIENTE_4" type="text" placeholder="Expediente 4">
          <input id="cdna_EXPEDIENTE_5" type="text" placeholder="Expediente 5">
        </div>
      </div>
    </div>
    <div class="field-wrap" data-scope="all">
      <div class="field">
        <label class="judicial-inline">
          <input id="cdna_INCLUIR_FIRMA" type="checkbox" checked />
          Incluir firma
        </label>
      </div>
    </div>
    """.strip()
    },

    "amonestacion": {
        "nombre": "AMONESTACIÓN",
        "template": CARTA_AMONESTACION,
        "order": [
            "code",
            "amonestado",
            "dni",
            "tipo",
            "dias",
            "cantidad",
            "super",
            "tiempo",
            "calculo",
            
        ],
        "labels": {
            "ciudad": "Ciudad",
            "fecha_documento": "Fecha del documento",
            "amonestado": "Amonestado",
            "dni": "DNI",
            "cargo_amonestado": "Cargo del amonestado",
            "cantidad": "Cantidad",
            "tipo": "Tipo",
            "dias": "Días",
            "calculo": "Cálculo del mes",
            "tiempo": "Tiempo acumulado",
            "super": "Supervisor firmante",
            "rol_destinatario": "Tipo de destinatario",
            "code": "N° de memorando",
        },
        "date_keys": set(),
        "select_keys": set(),
        "textarea_keys": set(),
        "defaults": {
            "ciudad": "Lima",
            "fecha_documento": "",
        },
        "extra_fields_top": """
        <div class="field-wrap" data-scope="all" style="grid-column: 1 / -1;">
          <div class="data-status" id="amonestacion_data_status">
            Data actualizada al ...
          </div>
        </div>
        <div class="field-wrap" data-scope="all">
          <div class="field">
            <label class="mini-label">Tipo de destinatario</label>
            <div class="switch-rol" id="amonestacion_switch_rol">
              <button type="button" class="switch-rol-btn active" data-value="asesor" id="btnRolAsesor">
                ASESOR
              </button>
              <button type="button" class="switch-rol-btn" data-value="supervisor" id="btnRolSupervisor">
                SUPERVISOR
              </button>
            </div>
            <input id="amonestacion_rol_destinatario" type="hidden" value="asesor">
          </div>
        </div>
        """.strip(),
        "extra_fields_bottom": """
        <div class="field-wrap" data-scope="all" style="grid-column: 1 / -1;">
          <div class="base-preview" id="amonestacion_base_preview">
            <div class="base-preview-title">Datos encontrados en base</div>
            <div class="base-preview-empty">Selecciona un amonestado para validar la información.</div>
          </div>
        </div>
        """.strip()
    }
}


NON_FIELDS = {"singular", "plural", "judicial", "asesor", "supervisor"}

def detectar_placeholders(template: str):
    raw_placeholders = set(re.findall(r"\{\{\s*\/?([A-Za-z0-9_]+)\s*\}\}", template))
    return sorted(p for p in raw_placeholders if p not in NON_FIELDS)

def extraer_placeholders_bloque(template: str, bloque: str):
    pattern = rf"\{{\{{\s*{bloque}\s*\}}\}}([\s\S]*?)\{{\{{\s*/{bloque}\s*\}}\}}"
    contenidos = re.findall(pattern, template)
    encontrados = set()
    for contenido in contenidos:
        encontrados.update(detectar_placeholders(contenido))
    return encontrados

def remover_bloques_condicionales(template: str, bloques=("asesor", "supervisor")):
    limpio = template
    for bloque in bloques:
        pattern = rf"\{{\{{\s*{bloque}\s*\}}\}}[\s\S]*?\{{\{{\s*/{bloque}\s*\}}\}}"
        limpio = re.sub(pattern, "", limpio)
    return limpio

def build_role_scopes(template: str):
    asesor_block = extraer_placeholders_bloque(template, "asesor")
    supervisor_block = extraer_placeholders_bloque(template, "supervisor")
    fuera_bloques = set(detectar_placeholders(remover_bloques_condicionales(template, ("asesor", "supervisor"))))

    todos = fuera_bloques | asesor_block | supervisor_block

    scopes = {}
    for key in todos:
        if key in fuera_bloques:
            scopes[key] = "all"
        elif key in asesor_block and key in supervisor_block:
            scopes[key] = "all"
        elif key in asesor_block:
            scopes[key] = "asesor"
        elif key in supervisor_block:
            scopes[key] = "supervisor"
        else:
            scopes[key] = "all"

    return scopes

def build_model_config(model_id: str, cfg: dict):
    detected = detectar_placeholders(cfg["template"])

    hidden_auto_fields = set()
    if model_id == "amonestacion":
        hidden_auto_fields = {"ciudad", "fecha_documento", "cargo_amonestado", "calculo"}
    elif model_id == "cdna":
        hidden_auto_fields = {"FECHA_HOY", "TIPO_DOC", "DETALLE_CREDITOS", "EXPEDIENTE"}

    ordered = [k for k in cfg["order"] if k not in hidden_auto_fields] + [
        k for k in detected if k not in cfg["order"] and k not in hidden_auto_fields
    ]

    role_scopes = {}
    if model_id == "amonestacion":
        role_scopes = build_role_scopes(cfg["template"])
        role_scopes = {k: v for k, v in role_scopes.items() if k not in hidden_auto_fields}
        for k in ordered:
            if k not in role_scopes:
                role_scopes[k] = "all"
    else:
        role_scopes = {k: "all" for k in ordered}

    return {
        "id": model_id,
        "nombre": cfg["nombre"],
        "template": cfg["template"],
        "placeholders": ordered,
        "labels": cfg["labels"],
        "date_keys": sorted(cfg["date_keys"]),
        "select_keys": sorted(cfg["select_keys"]),
        "textarea_keys": sorted(cfg["textarea_keys"]),
        "defaults": cfg["defaults"],
        "role_scopes": role_scopes,
        "extra_fields_top": cfg.get("extra_fields_top", cfg.get("extra_fields_html", "")),
        "extra_fields_bottom": cfg.get("extra_fields_bottom", ""),
    }

modelos = {model_id: build_model_config(model_id, cfg) for model_id, cfg in MODELOS.items()}

def wrap_field(scope: str, content: str) -> str:
    return f'<div class="field-wrap" data-scope="{scope}">{content}</div>'

def make_field(model_id: str, key: str, model_cfg: dict) -> str:
    labels = model_cfg["labels"]
    defaults = model_cfg["defaults"]
    date_keys = set(model_cfg["date_keys"])
    select_keys = set(model_cfg["select_keys"])
    textarea_keys = set(model_cfg["textarea_keys"])
    role_scopes = model_cfg.get("role_scopes", {})

    label = labels.get(key, key)
    default = defaults.get(key, "")
    field_id = f"{model_id}_{key}"
    scope = role_scopes.get(key, "all")

    if model_id == "cdna" and key == "CREDITOS":
        html = f"""
        <div class="field field-creditos">
          <div class="creditos-header">
            <label>Créditos</label>
    
            <div class="creditos-header-right">
              <div class="creditos-selector">
                <span>Cantidad</span>
                <select id="{model_id}_CREDITOS_COUNT" data-model="{model_id}">
                  <option value="1" selected>1</option>
                  <option value="2">2</option>
                  <option value="3">3</option>
                  <option value="4">4</option>
                  <option value="5">5</option>
                </select>
              </div>
    
              {'''
                <label class="judicial-inline">
                  <input id="cdna_JUDICIAL" type="checkbox" />
                  <span>Judicial</span>
                </label>
                ''' if model_id == "cdna" else ""}
            </div>
          </div>
    
          <div class="creditos-grid" id="{model_id}_CREDITOS_CONTAINER">
            <input id="{model_id}_CREDITOS_1" type="text" placeholder="Crédito 1">
            <input id="{model_id}_CREDITOS_2" type="text" placeholder="Crédito 2">
            <input id="{model_id}_CREDITOS_3" type="text" placeholder="Crédito 3">
            <input id="{model_id}_CREDITOS_4" type="text" placeholder="Crédito 4">
            <input id="{model_id}_CREDITOS_5" type="text" placeholder="Crédito 5">
          </div>
        </div>
        """.strip()
        return wrap_field(scope, html)

    if model_id == "amonestacion" and key == "amonestado":
        return f'''
        <div class="field-wrap" data-scope="{scope}" style="grid-column: 1 / -1;">
          <div class="field">
            <label for="{field_id}">{label}</label>
            <select id="{field_id}">
              <option value="">Selecciona...</option>
            </select>
          </div>
        </div>
        '''.strip()
        return wrap_field(scope, html)

    if model_id == "amonestacion" and key in {"dni", "super"}:
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <input id="{field_id}" type="text" value="{default}" readonly placeholder="{label}">
        </div>
        """.strip()
        return wrap_field(scope, html)

    if model_id == "amonestacion" and key == "cantidad":
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <input id="{field_id}" type="text" value="{default}" readonly placeholder="Se completa automáticamente">
        </div>
        """.strip()
        return wrap_field(scope, html)

    if model_id == "amonestacion" and key == "tipo":
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <select id="{field_id}">
            <option value="">Selecciona...</option>
            <option value="TARDANZA">TARDANZA</option>
            <option value="FALTA">FALTA</option>
          </select>
        </div>
        """.strip()
        return wrap_field(scope, html)

    if model_id == "amonestacion" and key == "dias":
        dias_botones = "\n".join(
            f'<button type="button" class="dia-btn" data-value="{i:02d}">{i}</button>'
            for i in range(1, 32)
        )
        html = f"""
        <div class="field field-dias">
          <label for="{field_id}">{label}</label>
          <input id="{field_id}" type="text" value="{default}" readonly placeholder="Selecciona uno o más días">
          <div class="dias-picker" id="{field_id}_picker">
            {dias_botones}
          </div>
        </div>
        """.strip()
        return wrap_field(scope, html)

    if model_id == "cdna" and key == "DOC":
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <input id="{field_id}" type="text" value="{default}" placeholder="Ingresa {label}">
        </div>
        """.strip()
        return wrap_field(scope, html)

    if key in select_keys:
        if key == "ENTIDAD":
            opts = ['<option value="">Selecciona una entidad...</option>']
            for opt in ENTIDAD_OPTIONS:
                selected = ' selected' if default and default == opt else ''
                opts.append(f'<option value="{opt}"{selected}>{opt}</option>')
            options_html = "\n".join(opts)
        else:
            options_html = f'<option value="{default}">{default}</option>' if default else '<option value="">Selecciona...</option>'

        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <select id="{field_id}">
            {options_html}
          </select>
        </div>
        """.strip()
        return wrap_field(scope, html)

    if key in textarea_keys:
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <textarea id="{field_id}" rows="3" placeholder="Ingresa {label}">{default}</textarea>
        </div>
        """.strip()
        return wrap_field(scope, html)

    if key in date_keys:
        html = f"""
        <div class="field">
          <label for="{field_id}">{label}</label>
          <input id="{field_id}" type="date" value="{default}">
        </div>
        """.strip()
        return wrap_field(scope, html)

    html = f"""
    <div class="field">
      <label for="{field_id}">{label}</label>
      <input id="{field_id}" type="text" value="{default}" placeholder="Ingresa {label}">
    </div>
    """.strip()
    return wrap_field(scope, html)

formularios_html = []
templates_html = []

for model_id, model_cfg in modelos.items():
    fields_html = "\n".join(
        make_field(model_id, k, model_cfg)
        for k in model_cfg["placeholders"]
    )

    top = model_cfg.get("extra_fields_top", "")
    bottom = model_cfg.get("extra_fields_bottom", "")
    
    if top:
        fields_html = top + "\n" + fields_html
    
    if bottom:
        fields_html = fields_html + "\n" + bottom

    active_class = "" if model_id == "amonestacion" else " hidden"

    formularios_html.append(f"""
    <div class="modelo-form{active_class}" id="formWrap_{model_id}" data-model="{model_id}">
      <form class="form-carta" data-model="{model_id}">
        <div class="grid">
          {fields_html}
        </div>

        <div class="actions">
          <button class="btn-primary" type="submit">GENERAR CARTA</button>
        </div>

        <p class="hint">Completa los campos y presiona <b>GENERAR CARTA</b> para ver la carta.</p>
      </form>
    </div>
    """.strip())

    template_safe = model_cfg["template"].replace("</template>", "<\\/template>")
    templates_html.append(f"""
    <template id="tplCarta_{model_id}">
{template_safe}
    </template>
    """.strip())

modelos_js = {}
for model_id, model_cfg in modelos.items():
    modelos_js[model_id] = {
        "id": model_id,
        "nombre": model_cfg["nombre"],
        "placeholders": model_cfg["placeholders"],
        "labels": model_cfg["labels"],
        "date_keys": model_cfg["date_keys"],
        "select_keys": model_cfg["select_keys"],
        "textarea_keys": model_cfg["textarea_keys"],
        "defaults": model_cfg["defaults"],
        "role_scopes": model_cfg["role_scopes"],
    }

modelos_json = json.dumps(modelos_js, ensure_ascii=False)

CSS_BASE = """
<style>
    body{
      margin:0;
      background:#f5f5f5;
      font-family: "Noto Sans", Arial, Helvetica, sans-serif;
      line-height:1.6;
      color:#222;
    }

    .hoja{
      width:21.59cm;
      min-height:27.94cm;
      height:27.94cm;
      margin:40px auto;
      background:#fff;
      padding:3.6cm 2cm 2.5cm;
      box-sizing:border-box;
      position:relative;
      overflow:visible;;
      border-radius:0;
      box-shadow:none;
    }

    .doc-header{
      position:absolute;
      top:1.2cm;
      left:2cm;
      right:2cm;
      margin:0;
    }

    .header-top.dual-logo{
      display:flex;
      justify-content:space-between;
      align-items:flex-start;
      width:100%;
    }
    
    .logo{
      max-width:150px;
      height:auto;
      display:block;
    }

    .fecha{
      text-align:right;
      white-space:nowrap;
      font-size:8.64pt;
      color:#555;
      letter-spacing:.2px;
      font-style: italic;
      margin-bottom:10px;
    }

    .titulo{
      text-align:center;
      margin:50px 0 26px;
      padding:10px 14px;
      font-weight:400;
      font-size:15pt;
    
      text-decoration: underline;
      text-underline-offset: 4px;
      text-decoration-thickness: 1px;
    }

    .saludo{
      margin:10px 0 18px;
      font-weight:700;
      font-size:10.5pt;
    }

    .cuerpo{
      font-size:9.75pt;
    }

    .cuerpo p{
      margin:0 0 14px;
      text-align:justify;
    }

    .despedida{
      margin-top:24px;
      font-size:10.5pt;
      text-align:right;
    }

    .firma{
      margin-top: 46px;
      text-align:center;
    }

    .firma-img{
      max-width: 190px;
      height:auto;
      display:block;
      margin: 0 auto 6px;
    }

    .firma-nombre{
      font-weight:700;
      font-size:11.5px;
      margin-top:2px;
    }

    .firma-cargo{
      font-size:11px;
      margin-top:2px;
    }

    .nota{
      position: absolute;
      left: 0;
      right: 0;
      bottom: 0;
    
      width: 100%;
      box-sizing: border-box;
    
      background: #f2f2f2;
      border-top: 1px solid #cfcfcf;
      border-bottom: 1px solid #cfcfcf;
    
      padding: 10px 2cm;
    }

    .nota small{
      display:block;
      margin:0;
      padding:0;
      border:0 !important;
      background:transparent !important;
      color:#333;
      font-size:8.25pt;
      line-height:1.4;
      text-align:justify;
    }

    .app-bar{
      max-width:780px;
      margin:40px auto 0;
      padding:0 12px;
      display:flex;
      justify-content:space-between;
      align-items:center;
      gap:12px;
    }

    .brand{
      font-weight:800;
      letter-spacing:.2px;
      color:#111;
    }

    .panel{
      max-width:780px;
      margin:12px auto 0;
      background:#fff;
      border-radius:10px;
      box-shadow:0 8px 20px rgba(0,0,0,.08);
      padding:20px 18px;
    }

    .tabs-cartas{
      display:flex;
      gap:10px;
      margin-bottom:16px;
      flex-wrap:wrap;
    }

    .tab-carta{
      border:1px solid #ddd;
      background:#f3f3f3;
      color:#222;
      padding:10px 14px;
      border-radius:10px;
      font-weight:700;
      cursor:pointer;
    }

    .tab-carta.active{
      background:#111;
      color:#fff;
      border-color:#111;
    }

    .grid{
      display:grid;
      grid-template-columns: 1fr 1fr;
      gap:12px;
      align-items:start;
    }

    @media (max-width: 760px){
      .grid{ grid-template-columns: 1fr; }
    }

    .field-wrap{
      display:block;
    }

    .field-wrap.hidden-scope{
      display:none !important;
    }

    .field label{
      display:block;
      font-size:12px;
      color:#444;
      margin-bottom:6px;
      font-weight:700;
    }

    .field input,
    .field textarea,
    .field select{
      width:100%;
      box-sizing:border-box;
      border:1px solid #ddd;
      border-radius:8px;
      padding:7px 12px;
      font-size:14px;
      outline:none;
    }

    .field input[readonly]{
      background:#f5f5f5;
      color:#444;
      cursor:not-allowed;
    }

    .doc-search-row{
      display:flex;
      gap:8px;
      align-items:center;
    }

    .doc-search-row input{
      flex:1;
      min-width:0;
    }

    .btn-buscar-bd{
      align-self:stretch;
      border:1px solid #111;
      background:#111;
      color:#fff;
      border-radius:8px;
      padding:7px 12px;
      font-size:12px;
      white-space:nowrap;
    }

    .bd-result-panel{
      margin-top:8px;
      border:1px solid #d9d9d9;
      border-radius:8px;
      background:#fafafa;
      padding:10px;
      font-size:12px;
      color:#222;
    }

    .bd-result-panel.hidden{
      display:none;
    }

    .bd-result-title{
      font-weight:800;
      margin-bottom:8px;
    }

    .bd-result-list{
      display:grid;
      gap:8px;
    }

    .bd-result-item{
      border-top:1px solid #e6e6e6;
      padding-top:8px;
    }

    .bd-result-item:first-child{
      border-top:0;
      padding-top:0;
    }

    .bd-result-grid{
      display:grid;
      grid-template-columns:max-content minmax(0, 1fr);
      gap:4px 8px;
    }

    .bd-result-label{
      color:#555;
      font-weight:800;
    }

    .bd-result-empty{
      color:#666;
    }

    .field-dias{
      grid-column: span 2;
    }

    .dias-picker{
      margin-top:10px;
      display:grid;
      grid-template-columns: repeat(7, minmax(0, 1fr));
      gap:8px;
    }

    .dia-btn{
      border:1px solid #d8d8d8;
      background:#f7f7f7;
      color:#222;
      padding:8px 0;
      border-radius:8px;
      font-size:13px;
      font-weight:700;
      cursor:pointer;
    }

    .dia-btn.activo{
      background:#111;
      color:#fff;
      border-color:#111;
    }

    @media (max-width: 760px){
      .dias-picker{
        grid-template-columns: repeat(5, minmax(0, 1fr));
      }
    }

    .mini-label{
      display:block;
      font-size:12px;
      color:#444;
      margin-bottom:6px;
      font-weight:700;
    }

    .creditos-grid{
      display:grid;
      grid-template-columns: 1fr 1fr;
      gap:10px;
    }

    @media (max-width: 760px){
      .creditos-grid{ grid-template-columns: 1fr; }
    }

    .creditos-header{
      display:flex;
      justify-content:space-between;
      align-items:center;
      margin-bottom:10px;
    }

    .creditos-header > label{
      font-size:12px;
      font-weight:700;
      color:#444;
    }

    .creditos-selector{
      display:flex;
      align-items:center;
      gap:6px;
      font-size:12px;
      color:#555;
    }

    .creditos-selector select{
      width:auto;
      padding:4px 8px;
      font-size:12px;
      border-radius:6px;
    }

    .creditos-header-right{
      display:flex;
      align-items:center;
      gap:14px;
      flex-wrap:wrap;
    }
    
    .judicial-inline{
      display:flex;
      align-items:center;
      gap:6px;
      font-size:12px;
      color:#444;
      font-weight:700;
      margin:0;
    }
    
    .judicial-inline input{
      width:auto !important;
      margin:0;
    }

    .switch-rol{
      display:flex;
      gap:8px;
      flex-wrap:wrap;
    }

    .switch-rol-btn{
      border:1px solid #d0d0d0;
      background:#f3f3f3;
      color:#222;
      padding:10px 16px;
      border-radius:10px;
      font-weight:800;
      cursor:pointer;
    }

    .switch-rol-btn.active{
      background:#111;
      color:#fff;
      border-color:#111;
    }

    .actions{
      display:flex;
      gap:10px;
      justify-content:flex-end;
      margin-top:14px;
      grid-column:1 / -1;
    }

    button{
      border:0;
      border-radius:10px;
      padding:10px 14px;
      font-weight:800;
      cursor:pointer;
    }

    .btn-primary{
      background:#111;
      color:#fff;
    }

    .btn-ghost{
      background:#eee;
      color:#111;
    }

    .btn-word{
      background:#1976d2; /* azul */
      color:#fff;
    }
    
    .btn-word:hover{
      background:#1565c0;
    }
    
    .btn-pdf{
      background:#e53935; /* rojo */
      color:#fff;
    }
    
    .btn-pdf:hover{
      background:#c62828;
    }

    .hint{
      font-size:12px;
      color:#666;
      margin:10px 0 0;
    }

    .hidden{
      display:none !important;
    }

    .data-status{
      border:1px solid #d8d8d8;
      background:#f7f7f7;
      color:#222;
      border-radius:8px;
      padding:10px 12px;
      font-size:12px;
      font-weight:800;
    }

    .base-preview{
      border:1px solid #d8d8d8;
      background:#fff;
      border-radius:8px;
      padding:12px;
      overflow:auto;
    }

    .base-preview-title{
      font-size:12px;
      font-weight:800;
      color:#222;
      margin-bottom:8px;
    }

    .base-preview-empty{
      color:#666;
      font-size:12px;
    }

    .base-preview table{
      width:100%;
      border-collapse:collapse;
      font-size:12px;
      min-width:560px;
    }

    .base-preview th,
    .base-preview td{
      border:1px solid #e0e0e0;
      padding:7px 8px;
      text-align:left;
      white-space:nowrap;
    }

    .base-preview th{
      background:#f3f3f3;
      font-weight:800;
      color:#222;
    }

    .base-preview .base-si{
      font-weight:800;
      color:#111;
    }

    /* ===== AMONESTACIÓN ===== */
    body.word-memo{
      font-family: Tahoma, Arial, sans-serif;
      font-size:10pt;
      line-height:1.15;
      color:#111;
    }

    body.word-memo .doc-header{
      position:static;
      top:auto;
      left:auto;
      right:auto;
      margin:0;
    }

    body.word-memo .header-top{
      display:block;
      margin:0 0 8pt 0;
    }

    body.word-memo .logo{
      width:132px !important;
      height:auto !important;
      display:block;
      margin:0;
    }

    .memo-wrap{
      font-size:10.5pt;
      color:#111;
    }

    .memo-codigo{
      text-align:center;
      font-weight:700;
      margin-top:40px;
      margin-bottom:34px;
    }

    .memo-fecha{
      text-align:left;
      margin-top: 100px;
      margin-bottom:26px;
    }

    .memo-meta{
      margin-bottom:22px;
    }

    .memo-meta p{
      margin:0 0 8px;
    }

    .memo-cuerpo p{
      margin:0 0 16px;
      text-align:justify;
      line-height: 1.19;
    }

    .memo-firma{
      text-align:left;
    }

    .firma-linea{
      width:200px;
      border-top:1px solid #222;
      margin:0 0 10px 0;
    }
    .firma-bloque div{
      line-height:1.4;
    }

    .memo-recepcion{
      margin-top:44px;
    }

    .memo-recepcion p{
      margin:0 0 10px;
    }

    .memo-codigo,
    .memo-fecha,
    .memo-meta,
    .memo-recepcion{
      font-family: Tahoma, sans-serif;
      font-size:10pt;
    }

    .memo-cuerpo,
    .memo-cuerpo p,
    .memo-firma,
    .memo-firma div{
      font-family: Calibri, sans-serif;
      font-size:11pt;
    }

    .firma-espacio{
      margin:0;
      line-height:1;
    }
    
    .firma-linea-texto{
      margin:0 0 8px 0;
      line-height:1;
    }

    .field.error input,
    .field.error select,
    .field.error textarea {
      border: 2px solid #e53935 !important;
      background: #fff5f5;
    }
    
    .field.error label {
      color: #e53935;
    }
    
    .field.error::after {
      content: "Campo obligatorio";
      display: block;
      font-size: 11px;
      color: #e53935;
      margin-top: 4px;
    }
</style>
""".strip()

HTML_BASE = r"""
<!doctype html>
<html lang="es">
<head>
  <link href="https://fonts.googleapis.com/css2?family=Noto+Sans:wght@400;700&display=swap" rel="stylesheet">
  <meta charset="utf-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1" />
  <title>QUICK_DOC</title>
  %%CSS%%
  <script src="https://cdn.jsdelivr.net/npm/html2canvas@1.4.1/dist/html2canvas.min.js"></script>
  <script src="https://cdn.jsdelivr.net/npm/jspdf@2.5.1/dist/jspdf.umd.min.js"></script>
</head>
<body>

  <div class="app-bar">
      <div class="brand">QUICK DOC</div>
      <div>
            <button class="btn-ghost" id="btnEditar" type="button">Editar datos</button>
            <button class="btn-word hidden" id="btnWord" type="button">EXPORTAR WORD</button>
            <button class="btn-pdf hidden" id="btnPdf" type="button">EXPORTAR PDF</button>
      </div>
  </div>

  <section class="panel" id="panelFormulario">
    <div class="tabs-cartas" id="tabsCartas">
        <button class="tab-carta" type="button" data-model="cdna">CARTA DE NO ADEUDO</button>
        <button class="tab-carta active" type="button" data-model="amonestacion">AMONESTACIÓN</button>
    </div>

    %%FORMULARIOS%%
  </section>

  <section id="panelCarta" class="hidden">
    <div id="renderCarta"></div>
  </section>

  %%TEMPLATES%%

<script>
  const MODELOS = %%MODELOS%%;
  const CATALOGO_AMONESTACION = %%CATALOGO_AMONESTACION%%;
  const PASSWORDS = %%PASSWORDS%%;
  const ACCESS_STATE = {
    cdna_unlocked: false
  };
  let activeModel = 'amonestacion';

  function formatIsoToSlashDate(iso) {
    const m = String(iso || '').match(/^(\d{4})-(\d{2})-(\d{2})$/);
    if (!m) return String(iso || '');
    return `${m[3]}/${m[2]}/${m[1]}`;
  }

  const panelFormulario = document.getElementById('panelFormulario');
  const panelCarta = document.getElementById('panelCarta');
  const renderCarta = document.getElementById('renderCarta');
  const btnEditar = document.getElementById('btnEditar');
  const btnWord = document.getElementById('btnWord');
  const btnPdf = document.getElementById('btnPdf');
  const MONTHS_ES = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Setiembre','Octubre','Noviembre','Diciembre'];

  function solicitarAccesoCDNA() {
    if (ACCESS_STATE.cdna_unlocked) {
      switchModel('cdna');
      return;
    }

    const claveEsperada = (PASSWORDS['CARTA_CDNA'] || '').trim();
    if (!claveEsperada) {
      alert('No se encontró una clave configurada para CARTA_CDNA.');
      return;
    }

    const ingresada = window.prompt('Ingrese la clave para acceder a CARTA DE NO ADEUDO:');
    if (ingresada === null) return;

    if (String(ingresada).trim() === claveEsperada) {
      ACCESS_STATE.cdna_unlocked = true;
      switchModel('cdna');
    } else {
      alert('Clave incorrecta.');
    }
  }

  function getFieldId(modelId, key) {
    return `${modelId}_${key}`;
  }

  function getTodayLongEs() {
    const d = new Date();
    return `${d.getDate()} de ${MONTHS_ES[d.getMonth()]} de ${d.getFullYear()}`;
  }

  function getAmonestacionCatalogByRol() {
    const rol = (document.getElementById('amonestacion_rol_destinatario')?.value || 'asesor').trim();
    return rol === 'supervisor'
      ? (CATALOGO_AMONESTACION.supervisor || [])
      : (CATALOGO_AMONESTACION.asesor || []);
  }

  function syncAmonestacionDataStatus() {
    const el = document.getElementById('amonestacion_data_status');
    if (!el) return;
    const fecha = CATALOGO_AMONESTACION.fecha_max_indicadores_larga
      || formatIsoToSlashDate(CATALOGO_AMONESTACION.fecha_max_indicadores);
    el.textContent = fecha
      ? `Data actualizada al ${fecha}`
      : 'Data actualizada al: sin fecha disponible';
  }

  function renderAmonestacionBasePreview() {
    const panel = document.getElementById('amonestacion_base_preview');
    if (!panel) return;

    const persona = getAmonestacionPersonaSeleccionada();
    const tipoBase = (document.getElementById('amonestacion_tipo')?.value || '').trim().toUpperCase();
    const registros = persona?.registros_base || [];

    if (!persona) {
      panel.innerHTML = `
        <div class="base-preview-title">Datos encontrados en base</div>
        <div class="base-preview-empty">Selecciona un amonestado para validar la información.</div>
      `;
      return;
    }

    let filtrados = registros;
    if (tipoBase === 'FALTA') {
      filtrados = registros.filter(row => row.falta === 'SI');
    } else if (tipoBase === 'TARDANZA') {
      filtrados = registros.filter(row => row.tardanza === 'SI');
    }

    if (!filtrados.length) {
      panel.innerHTML = `
        <div class="base-preview-title">Datos encontrados en base</div>
        <div class="base-preview-empty">Sin registros${tipoBase ? ` para ${escapeHtml(tipoBase)}` : ''} de ${escapeHtml(persona.nombre)}.</div>
      `;
      return;
    }

    const rowsHtml = filtrados.map(row => `
      <tr>
        <td>${escapeHtml(formatIsoToSlashDate(row.fecha))}</td>
        <td>${escapeHtml(row.dia)}</td>
        <td class="${row.falta === 'SI' ? 'base-si' : ''}">${escapeHtml(row.falta)}</td>
        <td class="${row.tardanza === 'SI' ? 'base-si' : ''}">${escapeHtml(row.tardanza)}</td>
        <td>${escapeHtml(formatTiempoHoraCorta(row.tiempo))}</td>
      </tr>
    `).join('');

    panel.innerHTML = `
      <div class="base-preview-title">Datos encontrados en base: ${escapeHtml(persona.nombre)}</div>
      <table>
        <thead>
          <tr>
            <th>Fecha</th>
            <th>Día</th>
            <th>Falta</th>
            <th>Tardanza</th>
            <th>Tiempo</th>
          </tr>
        </thead>
        <tbody>${rowsHtml}</tbody>
      </table>
    `;
  }

  function syncAmonestadoOptionsByRol() {
    const sel = document.getElementById('amonestacion_amonestado');
    const dni = document.getElementById('amonestacion_dni');
    const superInp = document.getElementById('amonestacion_super');
    if (!sel) return;

    const catalogo = getAmonestacionCatalogByRol();
    const valorActual = sel.value || '';

    sel.innerHTML = '<option value="">Selecciona...</option>' +
      catalogo.map(p => `<option value="${p.nombre}">${p.nombre}</option>`).join('');

    const existeActual = catalogo.some(p => p.nombre === valorActual);
    sel.value = existeActual ? valorActual : '';

    if (!existeActual) {
      if (dni) dni.value = '';
      if (superInp) superInp.value = '';
    }

    syncDatosAmonestacionDesdeSeleccion();
  }

  function syncCdnaExpedientesInputs() {
      const chkJudicial = document.getElementById('cdna_JUDICIAL');
      const wrap = document.getElementById('cdna_EXPEDIENTES_WRAP');
      const creditosCount = document.getElementById('cdna_CREDITOS_COUNT');
    
      if (!wrap || !creditosCount) return;
    
      const judicialActivo = !!chkJudicial?.checked;
      const n = Math.max(1, Math.min(5, Number(creditosCount.value || 1)));
    
      wrap.classList.toggle('hidden', !judicialActivo);
    
      for (let i = 1; i <= 5; i++) {
        const inp = document.getElementById(`cdna_EXPEDIENTE_${i}`);
        if (!inp) continue;
    
        const visible = judicialActivo && i <= n;
        inp.classList.toggle('hidden', !visible);
    
        if (!visible) inp.value = '';
      }
  }

  function getAmonestacionPersonaSeleccionada() {
    const sel = document.getElementById('amonestacion_amonestado');
    if (!sel) return null;

    const nombre = (sel.value || '').trim();
    const catalogo = getAmonestacionCatalogByRol();
    return catalogo.find(p => p.nombre === nombre) || null;
  }

  function syncDatosAmonestacionDesdeSeleccion() {
    const dni = document.getElementById('amonestacion_dni');
    const superInp = document.getElementById('amonestacion_super');
    const persona = getAmonestacionPersonaSeleccionada();

    if (!persona) {
      if (dni) dni.value = '';
      if (superInp) superInp.value = '';
      syncIncidenciasAmonestacionDesdeTipo();
      return;
    }

    if (dni) dni.value = persona.dni || '';
    if (superInp) superInp.value = persona.supervisor || '';
    syncIncidenciasAmonestacionDesdeTipo();
    renderAmonestacionBasePreview();
  }

  function setDiasAmonestacion(dias) {
    const input = document.getElementById('amonestacion_dias');
    if (!input) return;
    const limpios = [...new Set((dias || []).map(v => String(v).padStart(2, '0')))]
      .sort((a, b) => Number(a) - Number(b));
    input.value = limpios.join(', ');
    syncDiasButtonsFromInput();
    syncCantidadDesdeDias();
    syncTiempoTardanzaDesdeDias();
  }

  function syncIncidenciasAmonestacionDesdeTipo() {
    const persona = getAmonestacionPersonaSeleccionada();
    const tipoBase = (document.getElementById('amonestacion_tipo')?.value || '').trim().toUpperCase();
    const dias = tipoBase === 'FALTA'
      ? (persona?.faltas_dias || [])
      : tipoBase === 'TARDANZA'
        ? (persona?.tardanzas_dias || [])
        : [];

    setDiasAmonestacion(dias);
    renderAmonestacionBasePreview();
  }

  function syncDiasButtonsFromInput() {
    const input = document.getElementById('amonestacion_dias');
    if (!input) return;

    const seleccionados = new Set(
      (input.value || '')
        .split(',')
        .map(v => v.trim())
        .filter(Boolean)
    );

    document.querySelectorAll('#amonestacion_dias_picker .dia-btn').forEach(btn => {
      btn.classList.toggle('activo', seleccionados.has(btn.dataset.value));
    });
  }

  function getDiasAmonestacionSeleccionados() {
    return (document.getElementById('amonestacion_dias')?.value || '')
      .split(',')
      .map(v => v.trim())
      .filter(Boolean)
      .map(v => String(v).padStart(2, '0'));
  }

  function syncCantidadDesdeDias() {
    const cantidadInput = document.getElementById('amonestacion_cantidad');
    if (!cantidadInput) return;

    const total = getDiasAmonestacionSeleccionados().length;
    cantidadInput.value = total ? String(total) : '';
  }

  function parseTiempoExcelDias(value) {
    const n = Number(String(value || '').replace(',', '.'));
    return Number.isFinite(n) ? n : 0;
  }

  function formatTiempoHoraCorta(value) {
    const totalMinutos = Math.round(parseTiempoExcelDias(value) * 24 * 60);
    if (!totalMinutos) return '';
    const horas = Math.floor(totalMinutos / 60);
    const minutos = totalMinutos % 60;
    return `${String(horas).padStart(2, '0')}:${String(minutos).padStart(2, '0')} h`;
  }

  function formatTiempoTardanza(totalDias) {
    const totalMinutos = Math.round(totalDias * 24 * 60);
    if (!totalMinutos) return '';

    const horas = Math.floor(totalMinutos / 60);
    const minutos = totalMinutos % 60;
    const partes = [];
    if (horas) partes.push(`${horas} hora${horas === 1 ? '' : 's'}`);
    if (minutos) partes.push(`${minutos} minuto${minutos === 1 ? '' : 's'}`);
    return partes.join(' y ');
  }

  function syncTiempoTardanzaDesdeDias() {
    const tiempoInput = document.getElementById('amonestacion_tiempo');
    if (!tiempoInput) return;

    const tipoBase = (document.getElementById('amonestacion_tipo')?.value || '').trim().toUpperCase();
    const persona = getAmonestacionPersonaSeleccionada();
    if (tipoBase !== 'TARDANZA' || !persona) {
      tiempoInput.value = '';
      return;
    }

    const tiempos = persona.tardanza_tiempos || {};
    const total = getDiasAmonestacionSeleccionados()
      .reduce((acc, dia) => acc + parseTiempoExcelDias(tiempos[dia]), 0);
    tiempoInput.value = formatTiempoTardanza(total);
  }

  function getTipoTextoFinal() {
    const tipoBase = (document.getElementById('amonestacion_tipo')?.value || '').trim().toUpperCase();
    const cantidad = Number(document.getElementById('amonestacion_cantidad')?.value || 0);

    if (!tipoBase) return '';

    if (tipoBase === 'TARDANZA') {
      return cantidad === 1 ? 'TARDANZA INJUSTIFICADA' : 'TARDANZAS INJUSTIFICADAS';
    }

    if (tipoBase === 'FALTA') {
      return cantidad === 1 ? 'FALTA INJUSTIFICADA' : 'FALTAS INJUSTIFICADAS';
    }

    return tipoBase;
  }

  function toggleDiaSeleccion(value) {
    const input = document.getElementById('amonestacion_dias');
    if (!input) return;

    let valores = (input.value || '')
      .split(',')
      .map(v => v.trim())
      .filter(Boolean);

    if (valores.includes(value)) {
      valores = valores.filter(v => v !== value);
    } else {
      valores.push(value);
    }

    valores = [...new Set(valores)].sort((a, b) => Number(a) - Number(b));
    input.value = valores.join(', ');
    syncDiasButtonsFromInput();
    syncCantidadDesdeDias();
    syncTiempoTardanzaDesdeDias();
  }
  
  function setRolDestinatario(value) {
    const hidden = document.getElementById('amonestacion_rol_destinatario');
    const btnAsesor = document.getElementById('btnRolAsesor');
    const btnSupervisor = document.getElementById('btnRolSupervisor');

    if (!hidden) return;

    hidden.value = value;
    btnAsesor?.classList.toggle('active', value === 'asesor');
    btnSupervisor?.classList.toggle('active', value === 'supervisor');

    syncAmonestacionFieldsByRol();
    syncAmonestadoOptionsByRol();
  }

  function setCdnaTipoDoc(value) {
      const hidden = document.getElementById('cdna_TIPO_DOC');
      const btnDni = document.getElementById('btnDocDni');
      const btnRuc = document.getElementById('btnDocRuc');
    
      if (!hidden) return;
    
      hidden.value = value;
      btnDni?.classList.toggle('active', value === 'DNI');
      btnRuc?.classList.toggle('active', value === 'RUC');
  }

  function syncAmonestacionFieldsByRol() {
    const rol = (document.getElementById('amonestacion_rol_destinatario')?.value || 'asesor').trim();
    const formWrap = document.getElementById('formWrap_amonestacion');
    if (!formWrap) return;

    formWrap.querySelectorAll('.field-wrap').forEach(el => {
      const scope = el.dataset.scope || 'all';
      const visible = (scope === 'all') || (scope === rol);
      el.classList.toggle('hidden-scope', !visible);
    });
  }

  function showForm() {
    panelFormulario.classList.remove('hidden');
    panelCarta.classList.add('hidden');
    btnWord?.classList.add('hidden');
    btnPdf?.classList.add('hidden');
    window.scrollTo({ top: 0, behavior: 'smooth' });
  }
    
  function showCarta() {
    panelFormulario.classList.add('hidden');
    panelCarta.classList.remove('hidden');
    btnWord?.classList.remove('hidden');
    btnPdf?.classList.remove('hidden');
    window.scrollTo({ top: 0, behavior: 'smooth' });
  }

  function switchModel(modelId) {
    activeModel = modelId;

    document.querySelectorAll('.tab-carta').forEach(btn => {
      btn.classList.toggle('active', btn.dataset.model === modelId);
    });

    document.querySelectorAll('.modelo-form').forEach(box => {
      box.classList.toggle('hidden', box.dataset.model !== modelId);
    });

    syncCreditosInputs();
    syncCdnaJudicialUI();
    syncAmonestacionFieldsByRol();
    syncAmonestadoOptionsByRol();
    syncAmonestacionDataStatus();
    renderAmonestacionBasePreview();
    syncDiasButtonsFromInput();
    syncCantidadDesdeDias();
    syncTiempoTardanzaDesdeDias();
  }

  function syncCreditosInputs() {
      if (activeModel !== 'cdna') return;
    
      const creditosCount = document.getElementById('cdna_CREDITOS_COUNT');
      const creditosContainer = document.getElementById('cdna_CREDITOS_CONTAINER');
      if (!creditosCount || !creditosContainer) return;
    
      const n = Math.max(1, Math.min(5, Number(creditosCount.value || 1)));
      for (let i = 1; i <= 5; i++) {
        const inp = document.getElementById(`cdna_CREDITOS_${i}`);
        if (!inp) continue;
        const visible = i <= n;
        inp.classList.toggle('hidden', !visible);
        if (!visible) inp.value = '';
      }
    
      syncCdnaExpedientesInputs();
  }

  function escapeHtml(value) {
    return String(value ?? '').replace(/[&<>"']/g, ch => ({
      '&': '&amp;',
      '<': '&lt;',
      '>': '&gt;',
      '"': '&quot;',
      "'": '&#39;'
    }[ch]));
  }

    function normalizeDocKey(value) {
      return String(value || '').replace(/\\D/g, '').trim();
    }

  function excelDateToIso(value) {
    const raw = String(value || '').trim();
    if (!raw) return '';

    const iso = raw.match(/^(\d{4})-(\d{2})-(\d{2})/);
    if (iso) return `${iso[1]}-${iso[2]}-${iso[3]}`;

    const local = raw.match(/^(\d{1,2})[\/\-](\d{1,2})[\/\-](\d{2,4})$/);
    if (local) {
      const year = local[3].length === 2 ? `20${local[3]}` : local[3];
      return `${year}-${String(local[2]).padStart(2, '0')}-${String(local[1]).padStart(2, '0')}`;
    }

    return '';
  }

  function getBdRows() {
    return Array.isArray(BD_DOCUMENTOS?.rows) ? BD_DOCUMENTOS.rows : [];
  }

  function renderBdResultados(rows, doc) {
    const panel = document.getElementById('bdResultado');
    if (!panel) return;

    panel.classList.remove('hidden');

    if (!rows.length) {
      panel.innerHTML = `<div class="bd-result-empty">Sin resultados para ${escapeHtml(doc)}.</div>`;
      return;
    }

    const items = rows.map(row => `
      <div class="bd-result-item">
        <div class="bd-result-grid">
          <span class="bd-result-label">Documento</span><span>${escapeHtml(row.Documento_RUC)}</span>
          <span class="bd-result-label">Crédito</span><span>${escapeHtml(row.NumeroCuenta)}</span>
          <span class="bd-result-label">Fecha</span><span>${escapeHtml(row.FechaAdqui)}</span>
          <span class="bd-result-label">Entidad</span><span>${escapeHtml(row.ENTIDAD)}</span>
          <span class="bd-result-label">Cartera</span><span>${escapeHtml(row.cartera)}</span>
          <span class="bd-result-label">Saldo origen</span><span>${escapeHtml(row.SaldoTotalOrigen)}</span>
        </div>
      </div>
    `).join('');

    panel.innerHTML = `
      <div class="bd-result-title">${rows.length} resultado${rows.length === 1 ? '' : 's'} encontrado${rows.length === 1 ? '' : 's'}</div>
      <div class="bd-result-list">${items}</div>
    `;
  }

  function aplicarBdACamposCdna(rows) {
    if (!rows.length) return;

    const entidad = (rows[0].ENTIDAD || '').trim();
    const entidadSelect = document.getElementById('cdna_ENTIDAD');
    if (entidadSelect && [...entidadSelect.options].some(opt => opt.value === entidad)) {
      entidadSelect.value = entidad;
    }

    const fechaIso = excelDateToIso(rows[0].FechaAdqui);
    const fechaInput = document.getElementById('cdna_FECHA_CASTIGO');
    if (fechaInput && fechaIso) {
      fechaInput.value = fechaIso;
    }

    const creditos = rows
      .map(row => (row.NumeroCuenta || '').trim())
      .filter(Boolean);

    const creditosCount = document.getElementById('cdna_CREDITOS_COUNT');
    if (creditosCount && creditos.length) {
      creditosCount.value = String(Math.min(Math.max(creditos.length, 1), 5));
      syncCreditosInputs();
      creditos.slice(0, 5).forEach((credito, idx) => {
        const input = document.getElementById(`cdna_CREDITOS_${idx + 1}`);
        if (input) input.value = credito;
      });
    }
  }

  function buscarBDPorDocumento() {
    const doc = (document.getElementById('cdna_DOC')?.value || '').trim();
    const key = normalizeDocKey(doc);
    const panel = document.getElementById('bdResultado');

    if (!key) {
      if (panel) {
        panel.classList.remove('hidden');
        panel.innerHTML = '<div class="bd-result-empty">Ingresa un N° de documento para buscar.</div>';
      }
      return;
    }

    const rows = getBdRows().filter(row => normalizeDocKey(row.Documento_RUC) === key);
    renderBdResultados(rows, doc);
    aplicarBdACamposCdna(rows);
  }

  function syncCdnaJudicialUI() {
      syncCdnaExpedientesInputs();
  }

  function buildCreditosString() {
    const creditosCount = document.getElementById(`${activeModel}_CREDITOS_COUNT`);
    if (!creditosCount) return '';

    const n = Math.max(1, Math.min(5, Number(creditosCount.value || 1)));
    const vals = [];

    for (let i = 1; i <= n; i++) {
      const v = (document.getElementById(`${activeModel}_CREDITOS_${i}`)?.value || '').trim();
      if (v) vals.push(v);
    }

    return vals.join(', ');
  }

  function buildDetalleCreditos() {
      const creditosCount = document.getElementById('cdna_CREDITOS_COUNT');
      const judicialActivo = !!document.getElementById('cdna_JUDICIAL')?.checked;
    
      if (!creditosCount) return '';
    
      const n = Math.max(1, Math.min(5, Number(creditosCount.value || 1)));
      const detalles = [];
    
      for (let i = 1; i <= n; i++) {
        const credito = (document.getElementById(`cdna_CREDITOS_${i}`)?.value || '').trim();
        const expediente = (document.getElementById(`cdna_EXPEDIENTE_${i}`)?.value || '').trim();
    
        if (!credito) continue;
    
        if (judicialActivo) {
          detalles.push(`el crédito N.° ${credito}, anexado al expediente N.° ${expediente}`);
        } else {
          detalles.push(`el crédito N.° ${credito}`);
        }
      }
    
      if (!detalles.length) return '';
    
      if (detalles.length === 1) return detalles[0];
      if (detalles.length === 2) return `${detalles[0]} y ${detalles[1]}`;
    
      return `${detalles.slice(0, -1).join(', ')}, y ${detalles[detalles.length - 1]}`;
  }

  function applySingularPluralByCreditos(text) {
    const n = Number(document.getElementById('cdna_CREDITOS_COUNT')?.value || 1);
    const isSingular = (n === 1);
    const reSing = /\{\{\s*singular\s*\}\}([\s\S]*?)\{\{\s*\/singular\s*\}\}/gi;
    const rePlur = /\{\{\s*plural\s*\}\}([\s\S]*?)\{\{\s*\/plural\s*\}\}/gi;

    if (isSingular) {
      text = text.replace(reSing, '$1');
      text = text.replace(rePlur, '');
    } else {
      text = text.replace(rePlur, '$1');
      text = text.replace(reSing, '');
    }

    return text;
  }

  function applyJudicialBlock(text) {
    const on = !!document.getElementById('cdna_JUDICIAL')?.checked;
    const reJud = /\{\{\s*judicial\s*\}\}([\s\S]*?)\{\{\s*\/judicial\s*\}\}/gi;
    return on ? text.replace(reJud, '$1') : text.replace(reJud, '');
  }

  function applyFirmaBlock(text) {
    const incluir = document.getElementById('cdna_INCLUIR_FIRMA')?.checked !== false;
    const reFirma = /\{\{\s*firma\s*\}\}([\s\S]*?)\{\{\s*\/firma\s*\}\}/gi;
    return incluir ? text.replace(reFirma, '$1') : text.replace(reFirma, '');
  }

  function applyRolBlocks(text) {
    const rol = (document.getElementById('amonestacion_rol_destinatario')?.value || '').trim();
    const reAsesor = /\{\{\s*asesor\s*\}\}([\s\S]*?)\{\{\s*\/asesor\s*\}\}/gi;
    const reSupervisor = /\{\{\s*supervisor\s*\}\}([\s\S]*?)\{\{\s*\/supervisor\s*\}\}/gi;

    if (rol === 'asesor') {
      text = text.replace(reAsesor, '$1');
      text = text.replace(reSupervisor, '');
    } else if (rol === 'supervisor') {
      text = text.replace(reSupervisor, '$1');
      text = text.replace(reAsesor, '');
    } else {
      text = text.replace(reSupervisor, '');
      text = text.replace(reAsesor, '');
    }

    return text;
  }

  function applyTemplate(templateStr, data) {
    return templateStr.replace(/\{\{\s*([A-Za-z0-9_]+)\s*\}\}/g, (match, key) => {
      const v = data[key];
      return (v === undefined || v === null || v === '') ? match : String(v);
    });
  }

  function isoToLongEs(iso) {
    if (!iso) return '';
    const m = String(iso).match(/^(\\d{4})-(\\d{2})-(\\d{2})$/);
    if (!m) return '';
    const yyyy = Number(m[1]);
    const mm = Number(m[2]);
    const dd = Number(m[3]);
    if (!yyyy || mm < 1 || mm > 12 || dd < 1 || dd > 31) return '';
    return `${dd} de ${MONTHS_ES[mm - 1]} de ${yyyy}`;
  }

  function calcularPorcentajeMesActual() {
      const cantidad = Number(document.getElementById('amonestacion_cantidad')?.value || 0);
    
      const hoy = new Date();
      const año = hoy.getFullYear();
      const mes = hoy.getMonth(); // 0-11
    
      const diasDelMes = new Date(año, mes + 1, 0).getDate();
    
      if (!cantidad || !diasDelMes) return '';
    
      const porcentaje = (cantidad / diasDelMes) * 100;
      return `${porcentaje.toFixed(2)}%`;
  }

  function collectData(modelId) {
      const cfg = MODELOS[modelId];
      const data = {};
    
      for (const key of cfg.placeholders) {
        if (modelId === 'cdna' && key === 'CREDITOS') {
          data[key] = buildCreditosString();
          continue;
        }
    
        if (modelId === 'amonestacion' && key === 'tipo') {
          data[key] = getTipoTextoFinal();
          continue;
        }
    
        const el = document.getElementById(getFieldId(modelId, key));
        if (!el) continue;
    
        let v = (el.value || '').trim();
    
        if (cfg.date_keys.includes(key)) {
          v = isoToLongEs(v);
        }
    
        data[key] = v;
      }

      if (modelId === 'cdna') {
        data.TIPO_DOC = (document.getElementById('cdna_TIPO_DOC')?.value || 'DNI').trim();
        data.DOC = (document.getElementById('cdna_DOC')?.value || '').trim();
        data.FECHA_HOY = getTodayLongEs();
        data.DETALLE_CREDITOS = buildDetalleCreditos();
      }
    
      if (modelId === 'amonestacion') {
        const rol = (document.getElementById('amonestacion_rol_destinatario')?.value || 'asesor').trim();
    
        data.rol_destinatario = rol;
        data.ciudad = 'Lima';
        data.fecha_documento = getTodayLongEs();
        data.dni = (document.getElementById('amonestacion_dni')?.value || '').trim();
        data.super = (document.getElementById('amonestacion_super')?.value || '').trim();
        data.calculo = calcularPorcentajeMesActual();
    
        if (rol === 'asesor') {
          data.cargo_amonestado = 'Asesor de cobranza telefónica';
        } else if (rol === 'supervisor') {
          data.cargo_amonestado = 'Supervisor de cobranza telefónica';
        }
      }
    
      return data;
  }

  async function imageToBase64(img) {
      const src = img.currentSrc || img.src || img.getAttribute('src');
      if (!src) return null;
    
      if (src.startsWith('data:')) {
        return src;
      }

      try {
        if (!img.complete || !img.naturalWidth) {
          await img.decode();
        }

        const canvas = document.createElement('canvas');
        canvas.width = img.naturalWidth;
        canvas.height = img.naturalHeight;

        const ctx = canvas.getContext('2d');
        if (!ctx) throw new Error('No se pudo crear el contexto de imagen.');

        ctx.drawImage(img, 0, 0);
        return canvas.toDataURL('image/png');
      } catch (err) {
        console.warn('No se pudo convertir la imagen a base64; se conservará su ruta:', src, err);
        return src;
      }
  }
    
  async function clonarCartaConImagenesEmbebidas() {
      const original = document.querySelector('#renderCarta .hoja');
      if (!original) {
        throw new Error('No hay carta renderizada para exportar.');
      }
    
      const clon = original.cloneNode(true);
      const imgsOriginal = original.querySelectorAll('img');
      const imgsClon = clon.querySelectorAll('img');
    
      for (let i = 0; i < imgsOriginal.length; i++) {
        const base64 = await imageToBase64(imgsOriginal[i]);
    
        if (imgsClon[i]) {
          if (base64) imgsClon[i].setAttribute('src', base64);
    
          if (imgsClon[i].classList.contains('logo')) {
            imgsClon[i].setAttribute('width', '140');
            imgsClon[i].removeAttribute('height');
            imgsClon[i].style.width = '140px';
            imgsClon[i].style.height = 'auto';
            imgsClon[i].style.display = 'block';
          }
    
          if (imgsClon[i].classList.contains('firma-img')) {
            imgsClon[i].setAttribute('width', '180');
            imgsClon[i].removeAttribute('height');
            imgsClon[i].style.width = '180px';
            imgsClon[i].style.height = 'auto';
            imgsClon[i].style.display = 'block';
          }
        }
      }
    
      return clon;
  }
    
  function buildWordHtml(contentHtml) {
    const isMemoWord = activeModel === 'amonestacion';
    const pageCss = isMemoWord
      ? [
          '@page WordSection1 {',
          '  size: 21cm 29.7cm;',
          '  margin: 1.62cm 3cm 0.49cm 3cm;',
          '  mso-page-orientation: portrait;',
          '  mso-header-margin: 1.27cm;',
          '  mso-footer-margin: 1.27cm;',
          '}'
        ].join('\n')
      : [
          '@page WordSection1 {',
          '  size: 21.59cm 27.94cm;',
          '  margin: 0cm;',
          '  mso-page-orientation: portrait;',
          '  mso-header-margin: 0cm;',
          '  mso-footer-margin: 0cm;',
          '}'
        ].join('\n');

    const hojaCss = isMemoWord
      ? [
          '.hoja{',
          '  width:auto;',
          '  min-height:0;',
          '  height:auto;',
          '  margin:0;',
          '  background:#ffffff;',
          '  padding:0;',
          '  box-sizing:border-box;',
          '  position:static;',
          '  overflow:visible;',
          '  border-radius:0;',
          '  box-shadow:none;',
          '  page-break-after:auto;',
          '}'
        ].join('\n')
      : [
          '.hoja{',
          '  width:21.59cm;',
          '  min-height:27.94cm;',
          '  height:27.94cm;',
          '  margin:0 auto;',
          '  background:#fff;',
          '  padding:3.6cm 2cm 2.5cm;',
          '  box-sizing:border-box;',
          '  position:relative;',
          '  overflow:visible;',
          '  border-radius:0;',
          '  box-shadow:none;',
          '  page-break-after: avoid;',
          '}'
        ].join('\n');

    return `
    <!DOCTYPE html>
    <html xmlns:o="urn:schemas-microsoft-com:office:office"
      xmlns:w="urn:schemas-microsoft-com:office:word"
      xmlns="http://www.w3.org/TR/REC-html40">
    <head>
    <meta charset="utf-8">
    <title>Documento</title>
    <!--[if gte mso 9]>
    <xml>
      <w:WordDocument>
        <w:View>Print</w:View>
        <w:Zoom>100</w:Zoom>
        <w:DoNotOptimizeForBrowser/>
      </w:WordDocument>
      <w:LatentStyles DefLockedState="false" LatentStyleCount="156"/>
    </xml>
    <![endif]-->
    <style>
    ${pageCss}
    
    div.WordSection1 {
      page: WordSection1;
    }
    
    body{
      margin:0;
      background:#ffffff;
      font-family: "Noto Sans", Arial, Helvetica, sans-serif;
      font-size: 10pt;
      mso-fareast-font-family: Arial;
      mso-bidi-font-family: Arial;
      line-height:1.6;
      color:#222;
    }
    
    ${hojaCss}
    
    .doc-header{
      position:absolute;
      top:1.2cm;
      left:2cm;
      right:2cm;
      margin:0;
    }
    
    .header-top{
      display:flex;
      justify-content:flex-start;
      align-items:flex-start;
      margin-bottom:10px;
    }

    .dual-logo{
      display:flex;
      justify-content:space-between;
      align-items:flex-start;
    }
    
    .logo-entidad-wrap{
      display:flex;
      justify-content:flex-end;
      align-items:flex-start;
      flex:1;
    }
    
    .logo-entidad{
      max-width:140px;
      max-height:60px;
      object-fit:contain;
      display:block;
    }
    
    .fecha{
      text-align:right;
      white-space:nowrap;
      font-size:8.64pt;
      color:#555;
      letter-spacing:.2px;
      font-style: italic;
      margin-bottom:10px;
    }

    .titulo{
      text-align:center;
      margin:50px 0 26px;
      padding:10px 14px;
      font-weight:400;
      font-size:15pt;
      text-decoration: underline;
      text-underline-offset: 4px;
      text-decoration-thickness: 1px;
    }
    
    .saludo{
      margin:10px 0 18px;
      font-weight:700;
      font-size:10.5pt;
    }
    
    .cuerpo{
      font-size:9.75pt;
    }
    
    .cuerpo p{
      margin:0 0 14px;
      text-align:justify;
    }
    
    .despedida{
      margin-top:24px;
      font-size:10.5pt;
      text-align:right;
    }
    
    .firma{
      margin-top:46px;
      text-align:center;
    }
    
    .firma-img{
      width:180px;
      max-width:180px;
      height:auto;
      display:block;
      margin:0 auto 6px;
    }
    
    .firma-nombre{
      font-weight:700;
      font-size:11.5px;
      margin-top:2px;
    }
    
    .firma-cargo{
      font-size:11px;
      margin-top:2px;
    }

    .nota{
      position:absolute;
      left:0;
      right:0;
      bottom:0;
      width:100%;
      box-sizing:border-box;
      background:#f2f2f2;
      border-top:1px solid #cfcfcf;
      border-bottom:1px solid #cfcfcf;
      padding:10px 2cm;
    }
    
    .nota small{
      display:block;
      margin:0;
      padding:0;
      border:0 !important;
      background:transparent !important;
      color:#333;
      font-size:8.25pt;
      line-height:1.4;
      text-align:justify;
    }
    
    .memo-wrap{
      font-size:10.5pt;
      color:#111;
    }
    
    .memo-codigo{
      text-align:center;
      font-weight:700;
      margin-top:8pt;
      margin-bottom:34pt;
      font-family: Tahoma, sans-serif;
      font-size:10pt;
      mso-line-height-rule:exactly;
      line-height:12pt;
    }
    
    .memo-fecha{
      text-align:left;
      margin-top:0;
      margin-bottom:26pt;
      font-family: Tahoma, sans-serif;
      font-size:10pt;
      font-weight:700;
      mso-line-height-rule:exactly;
      line-height:12pt;
    }
    
    .memo-meta{
      margin-bottom:22pt;
      font-family: Tahoma, sans-serif;
      font-size:10pt;
    }
    
    .memo-meta p{
      margin:0 0 8pt;
      mso-line-height-rule:exactly;
      line-height:12pt;
    }
    
    .memo-cuerpo {
      font-family: Calibri, sans-serif !important;
      font-size: 11pt !important;
    }
    
    .memo-cuerpo p {
      font-family: Calibri, sans-serif !important;
      font-size: 11pt !important;
      line-height: 1.15 !important;
      margin:0 0 8pt !important;
      text-align:justify;
    }
    
    .memo-firma,
    .memo-firma div {
      font-family: Calibri, sans-serif !important;
      font-size: 11pt !important;
    }
    
    .memo-cuerpo p{
      margin:0 0 8pt;
      text-align:justify;
      line-height:1.15;
    }
    
    .memo-firma{
      text-align:left;
      font-family: Calibri, sans-serif !important;
      font-size:11pt !important;
    }
    
    .firma-espacio{
      margin:0;
      line-height:1;
      height:26pt;
    }
    
    .firma-linea-texto{
      margin:0 0 8px 0;
      font-family: Calibri, sans-serif !important;
      font-size:11pt !important;
      line-height:1;
    }
    
    .firma-bloque div{
      font-family: Calibri, sans-serif !important;
      font-size:11pt !important;
      line-height:1.2;
    }
    
    .memo-recepcion{
      margin-top:44pt;
      font-family: Tahoma, sans-serif;
      font-size:10pt;
    }
    
    .memo-recepcion p{
      margin:0 0 8pt;
      mso-line-height-rule:exactly;
      line-height:12pt;
    }

    .logo {
      width: 140px !important;
      height: auto !important;
    }
    
    </style>
    </head>
    <body class="${isMemoWord ? 'word-memo' : 'word-cdna'}">
      <div class="WordSection1">
        ${contentHtml}
      </div>
    </body>
    </html>
    `.trim();
  }
    
  function buildWordMhtml(contentHtml) {
    const boundary = '----=_NextPart_Codex_Document';
    const imageParts = [];
    let imageIndex = 0;

    const htmlConReferencias = contentHtml.replace(
      /src=(["'])data:([^;]+);base64,([^"']+)\1/g,
      (match, quote, mimeType, base64Data) => {
        imageIndex += 1;
        const extension = mimeType.includes('png') ? 'png' : 'jpg';
        const contentId = 'imagen' + imageIndex + '.' + extension;
        const base64ConSaltos = base64Data.replace(/(.{76})/g, '$1\r\n');

        imageParts.push(
          '--' + boundary + '\r\n' +
          'Content-Type: ' + mimeType + '\r\n' +
          'Content-Transfer-Encoding: base64\r\n' +
          'Content-Location: ' + contentId + '\r\n' +
          'Content-ID: <' + contentId + '>\r\n\r\n' +
          base64ConSaltos + '\r\n'
        );

        return 'src=' + quote + 'cid:' + contentId + quote;
      }
    );

    const htmlWord = buildWordHtml(htmlConReferencias);

    return (
      'MIME-Version: 1.0\r\n' +
      'Content-Type: multipart/related; boundary="' + boundary + '"\r\n\r\n' +
      '--' + boundary + '\r\n' +
      'Content-Type: text/html; charset="utf-8"\r\n' +
      'Content-Transfer-Encoding: 8bit\r\n\r\n' +
      htmlWord + '\r\n' +
      imageParts.join('') +
      '--' + boundary + '--'
    );
  }

  function descargarArchivo(nombre, contenido, mimeType, incluirBom = true) {
    const partes = incluirBom ? ['\ufeff', contenido] : [contenido];
    const blob = new Blob(partes, { type: mimeType });
    const url = URL.createObjectURL(blob);
    const a = document.createElement('a');
    a.href = url;
    a.download = nombre;
    document.body.appendChild(a);
    a.click();
    document.body.removeChild(a);
    URL.revokeObjectURL(url);
  }

  function buildNombreBaseDocumento() {
      const fecha = new Date();
      const año = fecha.getFullYear();
    
      if (activeModel === 'amonestacion') {
        const code = (document.getElementById('amonestacion_code')?.value || 'SIN-CODIGO').trim();
        const amonestado = (document.getElementById('amonestacion_amonestado')?.value || 'SIN-NOMBRE').trim();
    
        // 🔹 detectar tipo base (no el texto final largo)
        const tipoBase = (document.getElementById('amonestacion_tipo')?.value || '').trim().toUpperCase();
    
        let sigla = 'FT'; // default
    
        if (tipoBase === 'FALTA') {
          sigla = 'FI';
        } else if (tipoBase === 'TARDANZA') {
          sigla = 'FT';
        }
    
        return `${code}-${año} .Memorando Escrito ${sigla} ${amonestado}`;
      }

        if (activeModel === 'CAMP') {
          const nombre = (document.getElementById('CAMP_NOMBRE')?.value || 'DOCUMENTO')
            .trim()
            .replace(/[\\/:*?"<>|]+/g, '');
        
          const doc = (document.getElementById('CAMP_DOC')?.value || 'SIN-DOC').trim();
        
          return `CARTA-DE-CAMPAÑA_${doc}_${nombre}_${año}`;
        }

      if (activeModel === 'cdna') {
        const nombre = (document.getElementById('cdna_NOMBRE')?.value || 'DOCUMENTO')
          .trim()
          .replace(/[\\\\/:*?"<>|]+/g, '');
    
        return `CDNA_${nombre}_${año}`;
      }
    
      return `documento_${año}`;
  }
    
  function buildWordFileName() {
      return buildNombreBaseDocumento() + '.doc';
  }
    
  async function exportarWord() {
    try {
    const clon = await clonarCartaConImagenesEmbebidas();
    const htmlWord = buildWordMhtml(clon.outerHTML);
    const nombre = buildWordFileName();
    
    descargarArchivo(nombre, htmlWord, 'application/msword', false);
    } catch (err) {
    console.error(err);
    alert('No se pudo exportar el documento a Word.');
    }
  }

  function buildPdfFileName() {
      return buildNombreBaseDocumento() + '.pdf';
  }

  async function exportarPDF() {
      try {
        const original = document.querySelector('#renderCarta .hoja');
        if (!original) {
          throw new Error('No hay carta renderizada para exportar.');
        }
    
        const clon = await clonarCartaConImagenesEmbebidas();
    
        const contenedor = document.createElement('div');
        contenedor.style.position = 'fixed';
        contenedor.style.left = '-99999px';
        contenedor.style.top = '0';
        contenedor.style.zIndex = '-1';
        contenedor.style.background = '#ffffff';
        contenedor.style.padding = '0';
        contenedor.style.margin = '0';
    
        clon.style.margin = '0';
        clon.style.boxShadow = 'none';
        clon.style.borderRadius = '0';
        clon.style.background = '#ffffff';
    
        contenedor.appendChild(clon);
        document.body.appendChild(contenedor);
    
        const canvas = await html2canvas(clon, {
          scale: 2,
          useCORS: true,
          backgroundColor: '#ffffff',
          logging: false
        });
    
        document.body.removeChild(contenedor);
    
        const imgData = canvas.toDataURL('image/jpeg', 1.0);
    
        const { jsPDF } = window.jspdf;
        const pdf = new jsPDF({
          orientation: 'p',
          unit: 'mm',
          format: 'letter'
        });
    
        const pdfWidth = 215.9;
        const pdfHeight = 279.4;
    
        pdf.addImage(imgData, 'JPEG', 0, 0, pdfWidth, pdfHeight);
        pdf.save(buildPdfFileName());
    
      } catch (err) {
        console.error(err);
        alert('No se pudo exportar el documento a PDF.');
      }
  }

  document.querySelectorAll('.tab-carta').forEach(btn => {
      btn.addEventListener('click', () => {
        const modelId = btn.dataset.model;
    
        if (modelId === 'cdna') {
          solicitarAccesoCDNA();
          return;
        }
    
        switchModel(modelId);
      });
  });

  document.getElementById('btnRolAsesor')?.addEventListener('click', () => {
    setRolDestinatario('asesor');
  });

  document.getElementById('btnRolSupervisor')?.addEventListener('click', () => {
    setRolDestinatario('supervisor');
  });

  document.getElementById('btnDocDni')?.addEventListener('click', () => {
    setCdnaTipoDoc('DNI');
  });

  document.getElementById('btnDocRuc')?.addEventListener('click', () => {
    setCdnaTipoDoc('RUC');
  });

  btnEditar?.addEventListener('click', showForm);

  document.addEventListener('click', (e) => {
    if (e.target?.classList.contains('dia-btn')) {
      toggleDiaSeleccion(e.target.dataset.value);
    }
  });


  document.addEventListener('input', (e) => {
      if (
        e.target.id === 'amonestacion_code' ||
        e.target.id === 'amonestacion_dias'
      ) {
        const field = e.target.closest('.field');
        if (field) field.classList.remove('error');
      }
  });

  document.querySelectorAll('.form-carta').forEach(form => {
      form.addEventListener('submit', (e) => {
        e.preventDefault();
    
        const modelId = form.dataset.model;
    
        if (modelId === 'amonestacion') {
        
          const requeridos = [
            { id: 'amonestacion_code' },
            { id: 'amonestacion_amonestado' },
            { id: 'amonestacion_tipo' },
            { id: 'amonestacion_dias' },
          ];
        
          let hayErrores = false;
        
          // limpiar errores previos
          form.querySelectorAll('.field.error').forEach(el => {
            el.classList.remove('error');
          });
        
          requeridos.forEach(campo => {
            const el = document.getElementById(campo.id);
            const valor = (el?.value || '').trim();
        
            if (!valor) {
              hayErrores = true;
        
              const field = el.closest('.field');
              if (field) {
                field.classList.add('error');
              }
            }
          });
        
          if (hayErrores) {
            return; // ❌ NO genera la carta
          }
        }

        if (modelId === 'cdna') {
          const judicialActivo = !!document.getElementById('cdna_JUDICIAL')?.checked;
          const n = Math.max(1, Math.min(5, Number(document.getElementById('cdna_CREDITOS_COUNT')?.value || 1)));
        
          if (judicialActivo) {
            for (let i = 1; i <= n; i++) {
              const credito = (document.getElementById(`cdna_CREDITOS_${i}`)?.value || '').trim();
              const expediente = (document.getElementById(`cdna_EXPEDIENTE_${i}`)?.value || '').trim();
        
              if (credito && !expediente) {
                alert(`Falta completar el expediente ${i}.`);
                return;
              }
            }
          }
        }

        const data = collectData(modelId);
    
        if (modelId === 'cdna' && ('FECHA_HOY' in data) && !data.FECHA_HOY) {
          const d = new Date();
          const iso = `${d.getFullYear()}-${String(d.getMonth()+1).padStart(2,'0')}-${String(d.getDate()).padStart(2,'0')}`;
          data.FECHA_HOY = isoToLongEs(iso);
        }
    
        const tplRaw = document.getElementById(`tplCarta_${modelId}`).innerHTML;
        let tplFixed = tplRaw;
    
        if (modelId === 'cdna') {
          tplFixed = applySingularPluralByCreditos(tplFixed);
          tplFixed = applyJudicialBlock(tplFixed);
          tplFixed = applyFirmaBlock(tplFixed);
        }
if (modelId === 'amonestacion') {
          tplFixed = applyRolBlocks(tplFixed);
        }
    
        renderCarta.innerHTML = applyTemplate(tplFixed, data);
        showCarta();
      });
  });

  btnWord?.addEventListener('click', async () => {
      await exportarWord();
  });

  btnPdf?.addEventListener('click', async () => {
      await exportarPDF();
  });
    document.getElementById('cdna_CREDITOS_COUNT')?.addEventListener('change', syncCreditosInputs);
    document.getElementById('cdna_JUDICIAL')?.addEventListener('change', syncCdnaJudicialUI);
    const amonestadoSelect = document.getElementById('amonestacion_amonestado');
       if (amonestadoSelect) {
           amonestadoSelect.addEventListener('change', () => {
               syncDatosAmonestacionDesdeSeleccion();
               // Quitar clase de error si ya seleccionó uno
               amonestadoSelect.closest('.field')?.classList.remove('error');
           });
       } 
       document.getElementById('amonestacion_tipo')?.addEventListener('change', (e) => {
           e.target.closest('.field')?.classList.remove('error');
           syncIncidenciasAmonestacionDesdeTipo();
    });
    setRolDestinatario('asesor');
    syncAmonestacionDataStatus();
    renderAmonestacionBasePreview();
    setCdnaTipoDoc('DNI');
    syncAmonestadoOptionsByRol();
    syncDiasButtonsFromInput();
    syncCantidadDesdeDias();
    syncTiempoTardanzaDesdeDias();
    switchModel('amonestacion');
</script>

</body>
</html>
""".strip()

html_final = (
    HTML_BASE
    .replace("%%CSS%%", CSS_BASE)
    .replace("%%FORMULARIOS%%", "\n\n".join(formularios_html))
    .replace("%%TEMPLATES%%", "\n\n".join(templates_html))
    .replace("%%MODELOS%%", modelos_json)
    .replace("%%CATALOGO_AMONESTACION%%", CATALOGO_AMONESTACION_JSON)
    .replace("%%PASSWORDS%%", PASSWORDS_JSON)
)

ruta_index.write_text(html_final, encoding="utf-8")
print(f"✅ index.html actualizado correctamente en:\\n{ruta_index}")
print(f"📄 Excel usado para amonestación: {CATALOGO_AMONESTACION['excel_usado']} / hoja {CATALOGO_AMONESTACION['hoja_usada']}")
```
